## CART (Classification and Regression Tree)

### Decision Tree

Decision Tree merupakan model yang beroperasi dengan struktur berbentuk Binary Tree, di mana setiap node hanya dapat memiliki dua child node paling banyak. Dalam bidang data science, model ini diterapkan dalam konteks Supervised Learning, yaitu model dibentuk berdasarkan data training dan selanjutnya digunakan untuk membuat prediksi terhadap data yang disajikan. Data training yang digunakan akan diolah menjadi sebuah rangkaian aturan dalam bentuk Binary Tree. Aturan-aturan ini selanjutnya digunakan untuk proses prediksi pada dataset lain.

#### Training

Function di bawah ini merupakan function yang dapat digunakan untuk menghitung metrics hasil testing, baik untuk kasus regresi, maupun kasus klasifikasi.

```python
def regression_metrics(prediction, y_test):
    MAE = mean_absolute_error(y_test, prediction)
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    R_squared = r2_score(y_test, prediction)

    print('MAE: ' + str(MAE))
    print('MSE: ' + str(MSE))
    print('RMSE: ' + str(RMSE))
    print('R_squared: ' + str(R_squared))
```

```python
def classification_metrics(prediction, y_test):
  accuracy = accuracy_score(y_test, prediction)
  f1 = f1_score(y_test, prediction, average="macro")
  recall = recall_score(y_test, prediction, average="macro")
  precision = precision_score(y_test, prediction, average="macro")

  print('Accuracy: ' + str(accuracy))
  print('F1 Score: ' + str(f1))
  print('Recall Score: ' + str(recall))
  print('Precision Score: ' + str(precision))
```

#### Hyperparameter


Hyperparameter adalah parameter yang dapat kita sesuaikan atau konfigurasi sebelum memulai proses pembangunan model. Tujuan dari penyesuaian ini adalah untuk mengoptimalisasi performa model sehingga diharapkan dapat berfungsi lebih efektif.

> Ada beberapa jenis hyperparameter yang terdapat dalam beragam algoritma machine learning, termasuk dalam algoritma Decision Tree. Beberapa hyperparameter yang ada dalam algoritma Decision Tree yang disediakan oleh modul sklearn.tree antara lain:

- `criterion`: Metrik yang digunakan untuk mengevaluasi kualitas pemisahan node, contohnya:
  - `squared_error`: menggunakan reduksi varians sebagai kriteria seleksi fitur dan meminimalkan kerugian L2 dengan menggunakan mean dari setiap node terminal
  - `friedman_mse`: menghitung mean squared error dengan Friedman's improvement score untuk kemungkinan pemisahan (tidak perlu khawatir, ini tidak akan keluar di ujian)
  - `absolute_error`: meminimalkan kerugian L1 dengan menggunakan median dari setiap node terminal
  - `poisson`: memanfaatkan reduksi dalam Poisson deviance untuk menentukan pemisahan
- `max_depth`: Membatasi kedalaman maksimum dari pohon
- `min_samples_split`: Menentukan jumlah minimum sampel yang dibutuhkan untuk memecah sebuah node internal
Untuk informasi lebih detail, Anda bisa merujuk pada dokumentasi berikut:
[sklearn decision tree](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html)

Perubahan *hyperparameter* dapat mempengaruhi performa dari model. Terlihat bahwa MAE, MSE, RMSE, dan R_squared model Decision Tree dengan `max_depth = 3` berbeda dengan `max_depth = 6`.

### Ensemble Learning

*Ensemble Learning* adalah proses di mana kita menggunakan banyak model untuk mendapatkan hasil yang lebih baik dibandingkan dengan menggunakan satu model saja. Salah satu jenis *ensemble learning* adalah *bagging* (*bootstrap aggregating*).

Pada metode *bagging*, setiap model akan dilatih menggunakan subhimpunan dari data *training* (tidak keseluruhan data latih digunakan dalam setiap proses *training*). Pada subhimpunan data yang satu dengan yang lainnya bisa mengandung data yang sama (tidak harus beda semua), hal inilah yang disebut dengan *bootstrap sampling*. Nantinya, setiap model yang telah dilatih akan melakukan proses prediksi terhadap data yang baru. Hasil prediksi akhir didapatkan melalui sistem *majority voting* di mana hasil akhirnya akan menggunakan hasil prediksi yang paling banyak muncul pada model-model yang telah dibuat.

## Hyperparameter Tuning

Hyperparameter tuning dapat digunakan untuk mendapatkan hasil evaluasi yang terbaik untuk suatu model. Informasi lebih lanjut mengenai GridSearchCV dapat diakses [di sini](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)

Beberapa parameter yang sering digunakan, yaitu:
- `estimator`: model yang akan digunakan
- `param_grid`: kombinasi berbagai hyperparameter yang akan digunakan.
- `cv`: jumlah cross-validation yang ingin digunakan

### SOAL 3b

[2.5] Model Random Forest dianggap kurang dapat diinterpretasikan dibandingkan dengan Decision Tree. Apakah penyebab dari hasil tersebut?

Jawab:
Random Forest dianggap kurang dapat diinterpretasikan dibandingkan Decision Tree karena beberapa alasan:

- Kompleksitas Ensemble: Random Forest terdiri dari puluhan atau ratusan decision tree yang masing-masing dilatih pada subset data dan fitur yang berbeda. Sementara satu decision tree tunggal dapat dengan mudah divisualisasikan dan diinterpretasikan, visualisasi seluruh Random Forest tidak praktis dan sulit dipahami.
  
- Voting/Averaging Mechanism: Keputusan akhir Random Forest didasarkan pada hasil voting (klasifikasi) atau rata-rata (regresi) dari semua pohon, sehingga tidak ada jalur keputusan tunggal yang dapat diikuti seperti pada Decision Tree.

- Random Subspace Method: Setiap pohon dalam ensemble Random Forest dilatih menggunakan subset fitur yang berbeda (feature bagging), sehingga mengurangi kemampuan untuk melihat langsung pengaruh fitur tertentu terhadap hasil prediksi.

Referensi:

Breiman, L. (2001). Random Forests. Machine Learning, 45(1), 5-32. https://link.springer.com/article/10.1023/A:1010933404324
Scikit-learn Documentation, Random Forest: https://scikit-learn.org/stable/modules/ensemble.html#forest



### SOAL 3c

[2.5] Mekanisme voting dalam Random Forest Classifier adalah cara di mana model ini membuat prediksi akhir berdasarkan hasil dari beberapa Decision Tree yang membentuk ensemble-nya. Apakah yang membedakan mekanisme ini dengan mekanisme pengambil keputusan pada Decision Tree Classifier?

Jawab:
Perbedaan utama antara mekanisme pengambilan keputusan pada Random Forest Classifier dan Decision Tree Classifier adalah:

- Pendekatan Ensemble vs Singular: Random Forest menggunakan mekanisme voting dari banyak decision tree, sedangkan Decision Tree Classifier menggunakan satu decision tree saja untuk membuat prediksi.
  
- Cara Voting: Random Forest memprediksi kelas berdasarkan mayoritas hasil (voting) dari semua decision tree dalam ensemble. Setiap pohon memberikan satu suara, dan kelas dengan suara terbanyak menjadi hasil prediksi akhir.

- Robustness: Mekanisme voting membuat Random Forest lebih robust terhadap noise dan overfitting, karena kesalahan dari satu pohon dapat dikoreksi oleh pohon-pohon lainnya.

- Uncertainty Estimation: Dengan melihat distribusi voting, Random Forest dapat memberikan estimasi ketidakpastian prediksi (seberapa solid mayoritas voting), sementara Decision Tree individual hanya memberikan satu prediksi tanpa indikasi ketidakpastian.

Referensi:

Hastie, T., Tibshirani, R., & Friedman, J. (2009). The Elements of Statistical Learning: Data Mining, Inference, and Prediction. Springer Science & Business Media. https://web.stanford.edu/~hastie/ElemStatLearn/
Scikit-learn Documentation, Ensemble Methods: https://scikit-learn.org/stable/modules/ensemble.html

Jawab:

1. Bootstrapping dalam Random Forest
Bootstrapping dalam konteks Random Forest adalah teknik pengambilan sampel dengan pengembalian (sampling with replacement) dari dataset asli. Pada proses ini, setiap pohon keputusan (decision tree) dalam Random Forest dilatih menggunakan sampel data yang berbeda, meskipun berasal dari dataset yang sama.
Proses bootstrapping berkontribusi pada keragaman pohon keputusan dengan cara:

- Setiap pohon melihat subset data yang berbeda, sehingga masing-masing pohon memiliki "perspektif" yang sedikit berbeda tentang data
- Beberapa data mungkin muncul beberapa kali dalam satu sampel bootstrap, sementara data lain mungkin tidak terpilih sama sekali
- Keragaman ini membuat setiap pohon memiliki bias yang berbeda, dan ketika digabungkan, bias-bias ini cenderung saling meniadakan

Bootstrapping, dikombinasikan dengan pemilihan fitur acak pada setiap node, memastikan bahwa pohon-pohon dalam Random Forest tidak terlalu berkorelasi satu sama lain, yang penting untuk mengurangi overfitting dan meningkatkan generalisasi model.



2. Penggabungan Hasil Prediksi dalam Random Forest Regressor
Dalam Random Forest Regressor, prediksi dari setiap pohon digabungkan dengan cara yang sederhana namun efektif, yaitu dengan mengambil nilai rata-rata (mean) dari seluruh prediksi individu. Proses ini dapat dijabarkan sebagai berikut:

- Setiap pohon dalam ensemble memberikan prediksi numerik untuk sebuah input
- Random Forest mengumpulkan semua prediksi numerik ini
- Hasil akhir prediksi adalah rata-rata aritmatika dari semua prediksi individu

Pendekatan ini membantu mengurangi varians dan menghasilkan prediksi yang lebih stabil dibandingkan dengan menggunakan satu pohon keputusan saja.



3. Random Forest untuk Klasifikasi Multi-kelas
Random Forest dapat dengan mudah menangani masalah klasifikasi multi-kelas seperti pada prediksi Weather Type. Mekanismenya adalah sebagai berikut:

- Setiap pohon keputusan dalam Random Forest dilatih untuk memprediksi satu dari beberapa kategori cuaca (misalnya: sunny, rainy, snowy, dll)
- Ketika melakukan prediksi, setiap pohon "memberikan suara" untuk kategori cuaca tertentu
- Kelas yang mendapatkan suara terbanyak dari semua pohon (majority voting) akan menjadi prediksi akhir model
- Dalam kasus prediksi Weather Type, Random Forest dapat secara efektif membedakan berbagai kondisi cuaca berdasarkan fitur-fitur input seperti Temperature, Humidity, dan sebagainya

Pendekatan voting ini sangat efektif untuk klasifikasi multi-kelas karena memungkinkan model untuk menangkap berbagai aspek yang membedakan kelas-kelas tersebut.

Jawab:

1. Reproduksibilitas dalam Random Forest
Random Forest memang memasukkan unsur keacakan (randomness) melalui dua mekanisme utama:
- Bootstrapping (pengambilan sampel acak dengan pengembalian)
- Pemilihan fitur acak di setiap split node

Pengaruh terhadap reproduksibilitas:
- Hasil yang dihasilkan bisa berbeda-beda setiap kali model dijalankan
- Prediksi yang tidak konsisten antar pelatihan model
- Kesulitan dalam debugging karena hasil yang berubah-ubah

Untuk memastikan output yang konsisten, kita dapat:
- Menetapkan parameter random_state dengan nilai tetap pada konstruktor Random Forest (misal 2025)
- Menjaga kesamaan versi library yang digunakan (misal disini scikit-learn)
- Mendokumentasikan semua parameter model dan lingkungan eksekusi
- Menyimpan model yang sudah dilatih menggunakan pickle atau joblib daripada melatih ulang

Dengan langkah-langkah ini, kita dapat memastikan bahwa model Random Forest menghasilkan prediksi yang sama setiap kali dijalankan, yang sangat penting untuk aplikasi yang membutuhkan konsistensi dan untuk keperluan penelitian yang dapat direproduksi.


2. Ketahanan terhadap Outlier
Decision Tree dan Random Forest umumnya cukup robust (tahan) terhadap outlier karena beberapa alasan:

Decision Tree:
- Bekerja dengan mencari split optimal berdasarkan perbandingan nilai (lebih besar/lebih kecil), bukan jarak absolut
- Split node cenderung memisahkan outlier ke dalam node tersendiri
- Outlier hanya mempengaruhi cabang pohon tertentu, tidak seluruh model
- Namun, tree tunggal bisa over-fitting terhadap outlier jika pohon terlalu dalam


Random Forest:

- Jauh lebih robust karena mekanisme ensemble dan bootstrapping
- Setiap pohon hanya melihat subset data, sehingga outlier tidak muncul di semua pohon
- Proses voting mayoritas meredam pengaruh pohon-pohon yang terlalu dipengaruhi outlier
- Pemilihan fitur acak juga mengurangi dampak fitur dengan outlier ekstrem

Ketahanan terhadap outlier ini membuat Random Forest menjadi pilihan yang baik untuk dataset dunia nyata yang sering mengandung nilai-nilai tidak biasa atau pengukuran yang salah.

3. Kebutuhan Scaling dalam Random Forest
Random Forest tidak memerlukan scaling atau normalisasi fitur karena beberapa alasan:

- Random Forest bekerja berdasarkan aturan keputusan (decision rules) dan perbandingan nilai, bukan jarak atau magnitude
- Setiap split node membagi data berdasarkan threshold pada satu fitur, tanpa membandingkan antar fitur secara langsung
- Berbeda dengan algoritma berbasis jarak seperti KNN atau SVM yang sangat dipengaruhi skala fitur
- Kinerja Random Forest tetap konsisten terlepas dari apakah fiturnya memiliki skala 0-1 atau 0-1000

Sebagai ilustrasi sederhana, jika kita memiliki fitur "suhu" (dalam Â°C) dan "kelembaban" (dalam %), Random Forest akan memperlakukan kedua fitur ini sama pentingnya meskipun skalanya berbeda. Model akan mencari split terbaik di setiap fitur secara independen.
Ini membuat Random Forest lebih praktis digunakan karena mengurangi kebutuhan preprocessing data, dan lebih intuitif karena kita dapat menginterpretasi model langsung dalam skala asli fitur tanpa perlu mentransformasi kembali.
